```python
import numpy as np
import pandas as pd
import re 
import time
import requests
from bs4 import BeautifulSoup
import  matplotlib as plt
```

### Extracting the data from the url

```python
url="https://www.flipkart.com/search?q=water+purifier&as=on&as-show  =on&otracker=AS_Query_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=water+purifier&requestId=20975390-84a0-40ed-b96e-5f71c78955a0&page="
```

```python
page=requests.get(url)
pagecontent=page.text
soup=BeautifulSoup(pagecontent)
```

```python
page.status_code
```

```python
names=[]
ratings=[]
specifications=[]
prices=[]
original_costs=[]
discount_offers=[]
type_model=[]
delivery=[]
EMI_offers=[]
for i in range(1,127):
    url="https://www.flipkart.com/search?q=water+purifier&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_1_2_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_1_2_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=water+purifier&requestId=20975390-84a0-40ed-b96e-5f71c78955a0&page={}".format(i)

    page=requests.get(url)
    pagecontent=page.text
    soup=BeautifulSoup(pagecontent)
      
    for data in soup.find_all("div",class_="_3pLy-c row"):
        name=data.find("div",class_="_4rR01T")
        price=data.find("div",class_="_30jeq3 _1_WHN1")
        specification=data.find("ul",class_="_1xgFaf")
        rating=data.find("div",class_="_3LWZlK")
        original_cost=data.find("div",class_="_3I9_wc _27UcVY")
        discount=data.find("div",class_="_3Ay6Sb")
        delivery_offers=data.find("div",class_="_3tcB5a p8ucoS")
        emi=data.find("div",class_="_3xFhiH")
        types=data.find("li",class_="rgWa7D")
        if name is None:
            names.append(np.NaN)
        else:
            names.append(name.text)
            
        if price is None:
            prices.append(np.NaN)
        else:
            prices.append(price.text)
         
        if specification is None:
            specifications.append(np.NaN)
        else:
            specifications.append(specification.text)
            
        if rating is None:
            ratings.append(np.NaN)
        else:
            ratings.append(rating.text)
            
        if original_cost is None:
            original_costs.append(np.NaN)
        else:
            original_costs.append(original_cost.text)
            
        if discount is None:
            discount_offers.append(np.NaN)
        else:
            discount_offers.append(discount.text)
            
        if delivery_offers is None:
            delivery.append(np.NaN)
        else:
            delivery.append(delivery_offers.text)
            
        if emi is None:
            EMI_offers.append(np.NaN)
        else:
            EMI_offers.append(emi.text)
        if types is None:
            type_model.append(np.NaN)
        else:
            type_model.append(types.text)
```

```python
print(len(names))
print(len(ratings))
print(len(specifications))
print(len(prices))
print(len(original_costs))
print(len(discount_offers))
print(len(delivery))
print(len(EMI_offers))
print(len(type_model))
```

## Creation of the DataFrame

```python
waterpurifier_df=pd.DataFrame({"ProductName":names,"Specification":specifications,"Price":prices,"Original_cost":original_costs,"Rating":ratings,"Discount":discount_offers,"Delivery_offers":delivery,"EMI_offers":EMI_offers,"Type_models":type_model})
```
